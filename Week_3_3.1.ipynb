## 3.1	Prediction of the home prices using Multiple Linear Regression for a given dataset


```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('homeprices.csv')
df
```

```python
sns.lmplot(x='area',y='price',data=df)
```

```python
sns.lmplot(x='bedrooms',y='price',data=df)
```

```python
sns.lmplot(x='age',y='price',data=df)
```

```python
reg = LinearRegression()

reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
predicted = reg.predict([[3000,3,40]])
predicted
```

```python
reg.predict([[2500,4,5]])
```

```python
from sklearn.metrics import r2_score
y_original = df.price
y_original
```

```python
y_predicted = reg.predict(df[['area','bedrooms','age']])
y_predicted
```

```python
r_squ = r2_score(y_original,y_predicted)
r_squ
```
