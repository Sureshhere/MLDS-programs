```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("homeprices_Encoding.csv")
```

```python
df.head(10)
```

```python
le = LabelEncoder()
```

```python
df.town = le.fit_transform(df.town)
```

```python
df
```

```python
# retrieve training data
x = df[['town','area']]
x
```

```python
# retrieve target data
y = df.price
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(handle_unknown='ignore')
```

```python
x1 = ohe.fit_transform(df[['town']])
x1 = pd.DataFrame(x1.toarray())
x1
```

```python
# to avoid dummy variables trap, drop 0th column
x1 = x1.iloc[:,1:]
x1
```

```python
x = pd.concat([x,x1],axis='columns')
x
```

```python
# remove town as it is already encoded
x.drop('town',axis=1,inplace=True)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
x.columns=x.columns.astype(str) # making the columns: [1,2] as string ---> ['area','1','2']
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,1,0]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(x,y)
```
