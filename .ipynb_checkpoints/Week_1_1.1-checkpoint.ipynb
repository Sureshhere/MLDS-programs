## 1.1 : Implementation of Exploratory Data Analysis on CARS Data set

```python
import pandas as pd
import seaborn as sns
```

```python
cars = pd.read_csv("carsEDA.csv")
cars.head()
```

```python
cars.shape
```

```python
cars.columns
```

```python
cars.head(3)
```

```python
# remove unimportant cols: MSRP, Invoice 
cars = cars.drop(['MSRP', 'Invoice'], axis=1)
```

```python
cars.head(1)
```

```python
# remove duplicate rows if any 
#keep first row and remove other duplicate rows of that row 
cars = cars.drop_duplicates(keep='first')
```

```python
cars.shape # ROWS : 432--->429
```

```python
#To find the total no of rows which are missing value in a column 
cars.isnull().sum()
```

```python
cars.dropna(inplace=True)
```

```python
# sort the data w.r.t a column ---> here we sort on 'MPG_City’ in descending order 
cars_sort = cars.sort_values (by='MPG_City',ascending=False)
```

```python
cars_sort.head(5)
```

```python
cars.describe()
```

```python
cars.info()
```

```python
# find correlations between all columns 
cars.corr(method='pearson')
```

```python
# find correlations between all columns and 'EngineSize'.
cars.corr(method ='pearson')['EngineSize'] 
```
