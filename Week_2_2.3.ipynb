## 2.3	 Case Study: predict per capita income of Canada using Simple Linear Regression with train and test dataset.

Find out the per capita income of Canada during the years 2020 and 2021 based on the per capita income data provided using Simple Linear Regression Model.



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("canada_per_capita_income.csv")
dataset.head(3)
```

```python
x = dataset.iloc[:,0:1].values # retrieve 1st column as 2d array
y = dataset.iloc[:,-1].values # retrieve last column as 2d array
```

```python
# check the distribution of data is linear or not 
plt.scatter(x, y, color='red')
```

```python
#take 70% of data for training and 30% for testing 
#random_state indicates the random seed used in splitting the data 
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state = 5)
```

```python

# train the computer using Simple Linear Regression model 
from sklearn.linear_model import LinearRegression
reg = LinearRegression() 
reg.fit(x_train, y_train)
```

```python
# make prediction based on test data
y_pred = reg.predict(x_test)
y_pred
```

```python

#find the r squared value by comparing test data and predicted data 
from sklearn.metrics import r2_score 
r2_score (y_test, y_pred)
```

```python

# another way to know the score of a linear regression model 
reg.score (x_test, y_test)
```

```python

# predict the per capita income during the years 2020 and 2021 

# Output: array([41819.49650873, 42681.02869595])
# this means 41819$ in 2020 and 42681$ in 2021 years.

reg.predict([[2020], [2021]])
```

```python
# draw scatter plot and regression line # select below block and run at once
plt.scatter (x_train, y_train, color='red')
plt.plot(x_train, reg.predict(x_train), color= 'blue')
plt.title("PER CAPITA INCOME OF CANADA")
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.show()
```
