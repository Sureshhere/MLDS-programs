## 4.1	Apply Logistic Regression Model and predict whether a person takes insurance or not based on his age for a given dataset.


Given data like age and bought_insurance features. Apply Logistic Regression Model and predict whether a person takes insurance or not based on his age.  



```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('insurance_data.csv')
df.head()
```

```python
df.shape
```

```python
x = df.iloc[:,:1] # selecting 'age' column
y = df.iloc[:,1]  # selecting 'bought_insurance' column
```

```python
plt.scatter(x,y,marker='.',color='red')
```

```python
sns.regplot(x=x, y=y, data=df, logistic=True, ci=None)
```

```python
reg = LogisticRegression()
reg.fit(x,y)
```

```python
reg.score(x,y)
```

```python
reg.predict([[56]])
```

```python
reg.predict([[36]])
```
